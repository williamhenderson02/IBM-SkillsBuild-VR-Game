```python
#POST method is the format used for sending sql statements
conn = http.client.HTTPSConnection("bs2ipcul0apon0jufi80lite.db2.cloud.ibm.com")

#chamge 'commands' in the payload for other sql commands
payload = "{\"commands\":\"SELECT * FROM USERS\",\"limit\":10,\"separator\":\";\",\"stop_on_error\":\"yes\"}"

headers = {
    'content-type': "application/json",
    'authorization': "Bearer " + auth_token,
    'x-deployment-id': "crn:v1:bluemix:public:dashdb-for-transactions:eu-gb:a/1072bde6853044c2bcefefe05dfb8244:93c244ed-9e15-4e8e-8322-2a0d01d83ea5::"
    }

conn.request("POST", "/dbapi/v4/sql_jobs", payload, headers = headers)

res = conn.getresponse()
data = res.read()

print(data.decode("utf-8"))

data_decoded = json.loads(data.decode("utf-8"))

jobid = data_decoded["id"]
print(jobid)

request = "/dbapi/v4/sql_jobs/"
request = request + jobid
print(request)

#after getting the job id we can use a GET method to get the results of the sql statement
conn.request("GET", request, headers = headers)

res = conn.getresponse()
data = res.read()

print(data.decode("utf-8"))

data_decoded = json.loads(data.decode("utf-8"))

result = data_decoded["results"]
print(result)
columns = result[0]["columns"]
rows = result[0]['rows']

print(columns)
print(rows)
```
